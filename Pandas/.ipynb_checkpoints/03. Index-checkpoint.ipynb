# 3. Indx
데이터에 접근할 수 있는 주소 값

```python
import pandas as pd
# 딕셔너리 데이터
data = {
    # key : value
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
df=pd.DataFrame(data,index=['1번','2번','3번','4번','5번','6번','7번','8번'])
df
```

```python
# index 값 보여줌
df.index
```

## Index 이름 설정

```python
df.index.name='지원번호'
df
```

## Index 초기화

```python
# 새로운 인덱스 생기고, 원래 인덱스(지원번호)는 새로운 컬럼이 됨
df.reset_index()
```

```python
# 원래 인덱스(지원번호) 아예 삭제
# 실제 데이터에는 인덱스 삭제 된 것 반영 안 됨
df.reset_index(drop=True)
```

```python
df
```

```python
# df에서 인덱스 아예 삭제 -> inplace=True : 실제 데이터에 바로 반영
df.reset_index(drop=True,inplace=True)
```

```python
df
```

## Index 설정
지정한 column으로 Index를 설정

```python
# df.set_index('원하는 컬럼'), 실제 데이터에는 반영 안 됨
# df 안 써도 출력 뜨는 이유: df의 이름 컬럼을 인덱스로 하는 새로운 DataFrame을 반환
# 이 경우, 새로운 DataFrame이 반환되므로 바로 출력 결과가 Jupyter Notebook에서 자동으로 보이게 됨
df.set_index('이름')
```

```python
df
```

```python
# 실제 데이터에 반영
df.set_index('이름',inplace=True)
```

```python
df
```

## Index 정렬
Index를 기준으로 오름차순, 내림차순 정렬

```python
# 인덱스로 오름차순 정렬(asc)
df.sort_index()
```

```python
# ascending=False -> 내림차순으로 정렬(desc)
df.sort_index(ascending=False)
```
