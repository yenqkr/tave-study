# 15. 산점도 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.rcParams['font.family']='Malgun Gothic' # Window
matplotlib.rcParams['font.family']='AppleGothic' # Mac
matplotlib.rcParams['font.size']=15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus']=False # 한글 폰트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
import pandas as pd
df=pd.read_excel('../Pandas/score.xlsx')
df
```

```python
df['학년']=[3,3,2,1,1,3,2,2,]
df
```

```python
# 산점도 그래프
plt.scatter(df['영어'],df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
# 점크기 랜덤하게
import numpy as np
sizes=np.random.rand(8) * 1000
sizes
```

```python
# sizes 이용하여 점 크기 랜덤하게
plt.scatter(df['영어'],df['수학'],s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
# 학년별로 점 크기 다르게
sizes=df['학년'] * 500 # 1학년 500, 2학년 1000, 3학년 1500
plt.scatter(df['영어'],df['수학'],s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='viridis',alpha=0.3)
# cmap='viridis' 에서 c값에 따른 색 불러와 학년별 색이 지정됨(1,2,3에 해당하는 색 정해짐)
# alpha:투명도 설정
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='viridis',alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar()
# colorbar를 이용해 색이 무슨 숫자를 나타내는지 알 수 있음
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='viridis',alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks=[1,2,3],label='학년',shrink=0.5,orientation='horizontal') # colorbar 수정
# shrink: colorbar 크기 줄이기, orientation='horizontal': colorbar 가로로
```
