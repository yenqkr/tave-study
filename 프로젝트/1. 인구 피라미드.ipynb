# 1. 인구 피라미드

## 남자 데이터 정의

```python
import pandas as pd
df_m=pd.read_excel('202108_202108_연령별인구현황_월간.xlsx',skiprows=3,index_col='행정기관',usecols='B,E:Y')
# usecols='B,E:Y' : 가지고 오고 싶은 칼럼 선택
df_m.head(3)
```

```python
# df_m.iloc[0] # ,(콤마)가 있으면 문자로 인식 -> , 제거
# 1,195,951 -> 1195951(정수형)
df_m.iloc[0]=df_m.iloc[0].str.replace(',','').astype(int)
df_m
```

```python
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_w=pd.read_excel('201108_201108_연령별인구현황_월간.xlsx',skiprows=3,index_col='행정기관',usecols='B,AB:AV')
df_w.head(3)
```

```python
df_m.columns
```

```python
df_w.columns # 앞에 같은 컬럼명들이 있어서 뒤에 .1이 붙는다
```

```python
df_w.columns=df_m.columns # 컬럼명 통일
df_w.columns
```

```python
df_w.iloc[0]=df_w.iloc[0].str.replace(',','').astype(int)
df_w
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.rcParams['font.family']='Malgun Gothic' # Window
matplotlib.rcParams['font.family']='AppleGothic' # Mac
matplotlib.rcParams['font.size']=15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus']=False # 한글 폰트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns,-df_m.iloc[0]//1000) # 전국 남자 데이터, 단위 천명
plt.barh(df_w.columns,df_w.iloc[0]//1000) # 전국 여자 데이터, 단위 천명
plt.title('2021 대한민국 인구 피라미드')
plt.savefig('2021_인구피라미드.png',dpi=100)
plt.show()
```
